```python
import pandas as pd

file_path = './exam_sample.csv'

df = pd.read_csv(file_path, engine='python', encoding='CP949')
df
```

```python
df1=df.groupby(['class'])
df1
```

```python
# 반 중 A반 그룹만 화깅
df1.get_group('A')
```

```python
df.groupby('class').mean()
```

```python
# 반별, 성별 그룹 평균 확인
df.groupby(['class', 'sex']).mean()
```

```python
# 반별 수학 평균
df['math'].groupby(df['class']).mean()
df.groupby(df['class'])['math'].mean()
```

```python
# 반별 수학 개수
df['math'].groupby(df['class']).count()
```

```python
# 성별 수학 평균
df_mean=df['math'].groupby(df['sex']).mean()

sexgroup = df.groupby('sex')

print(df_mean)
print(sexgroup)
print(sexgroup.groups)
```

```python
# 남학생 수학평균
male=sexgroup.get_group('m')
subset = male[['sex', 'math']]
print(male)
print(subset)
```

### 데이터 구간화

```python
import pandas as pd

file_path = './exam_sample.csv'

df = pd.read_csv(file_path, engine='python', encoding='CP949')
df
```

```python
df1=pd.cut(df.math, 3)
```

```python
df1
```

```python
df2 = df.math.groupby(df1)
df2
df3=df2.agg(['count', 'mean', 'std', 'min', 'max'])
df3
```

```python
df = pd.get_dummies(df)
df
```

```python
# 전치 자료
df.T
```
